```python
import numpy as np
import tensorflow
from tensorflow import keras
from tensorflow.keras.applications import imagenet_utils
import os
```

```python
mobile = tensorflow.keras.applications.mobilenet.MobileNet()
mobile.summary()
```

```python
from tensorflow.keras.preprocessing import image
```

```python
import os 
os.chdir(r"C:\Users\ASUS\OneDrive\Images")
def wjedliyaimage(file):
    img_path = r'C:/Users/ASUS/OneDrive/Images/'
    img = image.load_img(img_path + file, target_size=(244,244))
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array ,axis=0) #ppour garder les memes dimension
    return tensorflow.keras.applications.mobilenet.preprocess_input(img_array_expanded_dims)
```

```python
from IPython.display import Image
Image(filename='C:/Users/ASUS/OneDrive/Images/1.jpg', height = 450 , width =300)
```

```python
preprocessed_image = wjedliyaimage("1.jpg")
predictions = mobile.predict(preprocessed_image)
#predictions
results = imagenet_utils.decode_predictions(predictions)
results
```
